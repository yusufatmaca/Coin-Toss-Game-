```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def coinTossFunction(x):
    coinToss = ["heads", "tails"]
    counterForHeads = 0
    counterForTails = 0
    for i in range (x):
        if (random.randint(0, 1) == 0):
            print(coinToss[0] + "!")
            counterForHeads += 1
        else:
            print(coinToss[1] + "!")
            counterForTails += 1

    A = []
    A.append(random.choice(coinToss))
    B = np.array(A)

    x = [coinToss[0], coinToss[1]]
    y = [counterForHeads,counterForTails]
    ratioOfHeads = counterForHeads / (counterForHeads + counterForTails)
    ratioOfTails = 1 - ratioOfHeads
    fig = plt.figure(figsize = (5, 10))
    plt.bar(x,y, width= 0.8)

    print("Total number of shot: ", (counterForHeads + counterForTails))
    print("Total number of heads: ", counterForHeads)
    print("Total number of tails: ", counterForTails)

coinTossFunction(1000)
```
